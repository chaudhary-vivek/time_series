```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
```

```python
df_ = pd.read_csv('Walmart.csv')
df_['Date'] = pd.to_datetime(df_['Date'], format='%d-%m-%Y')
```

```python
df_.groupby(['Store'])['Weekly_Sales'].sum().reset_index().sort_values(by = ['Weekly_Sales'], ascending = False).head(5)
```

```python
summary = df_.groupby('Store').agg({'Date': ['min', 'max', 'nunique']})
summary.columns = ['First_Date', 'Last_Date', 'Num_Weeks']
summary.head(5)
```

```python
df = df_[df_['Store'] == 20]
df = df.sort_values(by = 'Date')
print(df.shape)
df.head(5)
```

```python
fig, ax = plt.subplots(3, 2, figsize=(12, 8))
metrics = ['Weekly_Sales', 'Holiday_Flag', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']

for i, metric in enumerate(metrics):
    df.plot(x='Date', y=metric, ax=ax[i//2, i%2])
    ax[i//2, i%2].tick_params(axis='x', rotation=45)
    ax[i//2, i%2].set_title(metric)

plt.tight_layout()
plt.show()
```

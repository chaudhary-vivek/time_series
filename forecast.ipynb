```python
from data_preprocessing import data_load, forecast_arima, plot_comparison, plot_accuracy, forecast_lstm, train_model
import pandas as pd
```

```python
FORECAST_DATE = '2012-06-29'
DATE_COL = 'Date'
PATH = ''
FORECAST_PERIODS = 17
Y_VAR = "Weekly_Sales"
KEY_COL = 'Store'
```

```python
train, test = data_load(PATH, FORECAST_DATE, DATE_COL, Y_VAR)
```

```python
arima_forecasts = train_model(train, 'arima', KEY_COL ,DATE_COL, Y_VAR, FORECAST_PERIODS)
lstm_forecasts = train_model(train, 'lstm', KEY_COL ,DATE_COL, Y_VAR, FORECAST_PERIODS)
```

```python
plot_accuracy(arima_forecasts, test, KEY_COL, DATE_COL, Y_VAR, 'Arima')
plot_accuracy(lstm_forecasts, test, KEY_COL, DATE_COL, Y_VAR, 'LSTM')
```
